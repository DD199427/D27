```python
import nltk
```

```python
sentence = [("the", "DT"), ("little", "JJ"), ("yellow", "JJ"),("dog", "NN"), ("barked", "VBD"), ("at", "IN"),  ("the", "DT"), ("cat", "NN")]
```

```python
grammar="NP: {<DT>?<JJ>*<NN>}"
```

```python
cp=nltk.RegexpParser(grammar)
```

```python
result=cp.parse(sentence)
print(result)
```

```python
result.draw()
```

```python
doc=r"C:\Users\ADITYA\Desktop\Winsem\NLP\LoR Sample 1.txt"
raw_text = open(doc, 'r').read()
```

```python
sentences = nltk.sent_tokenize(raw_text)
```

```python
sentences = [nltk.word_tokenize(sent) for sent in sentences]
```

```python
sentences = [nltk.pos_tag(sent) for sent in sentences]
```

```python
print(sentences)
```

```python
grammar = "PRP:{<DT>?<IN>?<JJ>*<NN>}"
cp = nltk.RegexpParser(grammar)
result =[]
for i in range(len(sentences)):
    result.append(cp.parse(sentences[i]))
for r in result:
    print(r)
```

```python
for r in result:
    r.draw()
```
