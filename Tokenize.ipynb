```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
from nltk import word_tokenize
from urllib import request
```

```python
url="https://www.gutenberg.org/ebooks/42671.txt.utf-8"
response=request.urlopen(url)
rawtext=response.read().decode('utf8')
type(rawtext)
```

```python
len(rawtext)
#rawtext[:725]
```

```python
token=word_tokenize(rawtext)
```

```python
len(token)
```

```python
print(token)
```
